## Matplotlib example workflow

```python
# 0.import matplotlib and get it ready for plotting in jupyter
%matplotlib inline   
import matplotlib.pyplot as plt

# 1. Prepare data
x=[1,2,3,4]
y=[10,24,30,54]

# 2. Setup plot
fig,ax=plt.subplots(figsize=(5,5))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="simple plot",xlabel="x-axis",ylabel="y-axis")

#5.save & show(you save the whole figure)
fig.savefig("images/sample-plot.png")
```

### Bar plot

```python
# make a plot from dictionary
nut_butter_price = {"Almond butter":10,
                    "Peanut butter":8,
                    "Cashew butter":12}
fig,ax=plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values());
ax.set(title="butter shop",xlabel="Butter",ylabel="Price ($)");
```

```python
fig,ax=plt.subplots()
ax.barh(x,np.sin(x));
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()));
```

### Histogram

```python
x= np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

### subplots - two options

```python
# option 1 (recommended)
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot to each differnt axis 
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_price.keys(),nut_butter_price.values())
ax4.hist(x);
```

```python
# option2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_price.keys(),nut_butter_price.values())
ax[1,1].hist(np.random.randn(1000));
```

### which one should you use? (pyplot vs matplotlib oo method)

* when plotting something quickly, okay to use the pyplot method(plot.)
* when plotting something more advanced, use the oo method

### Create a subplot with chol & thalach which shares age(x-axis)

```python
over_50.head()
```

```python
# subplot creation
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

## cholestrol subplot(ax0);
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(xlabel="age",ylabel="cholestrol",
        title="Heart disease and cholestrol level")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50["chol"].mean(),color="orange",linestyle="--");

## thalach subplot(ax1);
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(xlabel="age",ylabel="max heart rate",title="Heart disease and max heart rate")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(),color="red",linestyle="--");
```

```python
# subplot creation
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

## cholestrol subplot(ax0);
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(ylabel="cholestrol",title="Heart disease and cholestrol level")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50["chol"].mean(),color="orange",linestyle="--");

## thalach subplot(ax1);
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(xlabel="age",ylabel="max heart rate",
        title="Heart disease and max heart rate")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(),linestyle="--");

# add a title to a figure:
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
# customizing the y and x axis limitations
# subplot creation
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)
## cholestrol subplot(ax0);
# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")
# customize ax0
ax0.set(ylabel="cholestrol",
        title="Heart disease and cholestrol level")
# change the x axis limits
ax0.set_xlim([50,80])
# Add a legend to ax0 
ax0.legend(*scatter.legend_elements(),  title="Target")
# Add a mean line
ax0.axhline(y=over_50["chol"].mean(),
            color="red",
            linestyle="--");
## thalach subplot(ax1);
# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")
# Customize ax1
ax1.set(xlabel="age",
        ylabel="max heart rate",
        title="Heart disease and max heart rate")
# Change ax1 x&y axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),   title="Target")
# Add a meanline
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--",
            color="red");
# add a title to a figure:
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
